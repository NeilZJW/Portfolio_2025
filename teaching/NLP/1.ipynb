## Character set

- [abc] Specifies the containing character
- [a-zA-Z] Specifies the cases of all English letters
- [^a-zA-Z] Specifies that all English letters should not be matched

```python
input2
```

```python
pattern = re.compile("[abc]")
re.findall(pattern, input2)
```

```python
pattern = re.compile("[a-zA-Z]")
re.findall(pattern, input2)
```

```python
pattern = re.compile("[^a-zA-Z]")
re.findall(pattern, input2)
```

## OR - Alternation

Two rules are juxtaposed and connected with "|" to indicate that one of them nedds to be matched.

[A-Za-z]|[0-9] means that a match can be made if it satisfies a number or letter. This rule is equivalent to [A-Za-z0-9]

```python
pattern = re.compile(r"[A-Z]|[0-9]")
re.findall(pattern, input2)
```

# Repetition

Regular Expressions can match strings of variable length

## "*" means 0 or multiple matches

```python
input2
```

```python
pattern = re.compile(r"\d*")
re.findall(pattern, input2)
```

## "+" means 1 or multiple matches

```python
pattern = re.compile(r"\d+")
re.findall(pattern, input2)
```

## "?" means 0 or 1 match

```python
pattern = re.compile(r"\d?")
re.findall(pattern, input2)
```

# Exact matching and minimal matching

## {m} matches exactly m times

```python
input2
```

```python
pattern = re.compile(r"\d{3}")
re.findall(pattern, input2)
```

## {m,n} matches at least m times and at most n times. (n > m)

```python
pattern = re.compile(r"\d{1,3}")
re.findall(pattern, input2)
```

# Lookahead

Lookahead is an advanced feature of Regular Expressions that checks if a condition is met when a match is made, but doesn't actually consume characters

## X(?=Y)

Matching X but only it is followed by Y

```python
pattern = r"abc(?=\d)"

str1 = "abc123"
str2 = "abcXYZ"

print(re.findall(pattern, str1))
print(re.findall(pattern, str2))
```

## X(?!Y)

Matching X but only if it is NOT followed by Y

```python
pattern = r"abc(?!\d)"

str1 = "abc123"
str2 = "abcXYZ"

print(re.findall(pattern, str1))
print(re.findall(pattern, str2))
```

## (?<=Y)X

Matching X if there is Y before it

```python
pattern = r"(?<=\d)abc"

str1 = "abc123"
str2 = "123abc"

print(re.findall(pattern, str1))
print(re.findall(pattern, str2))
```

## (?<!Y)X

Matching X if there is NO Y before it

```python
pattern = r"(?<!\d)abc"

str1 = "abc123"
str2 = "123abc"

print(re.findall(pattern, str1))
print(re.findall(pattern, str2))
```

# Match & Search

Instead of returning a simple list of strings, they return a MatchObject that gives you more information.

***If the match is unsuccessful, they return a NoneType. So before you can operate on the matching results, you need to check whether the match was successful.***

- match - Starting at the beginning of the string and fails if there is no match at the beginning.
- search - Skipping the beginning and continuing looking for a match

```python
input3 = "123 Natural Language Processing"
input4 = "Natural Language Processing 123"
```

## match

```python
 pattern = re.compile(r"\d")
 match3 = re.match(pattern, input3)
 match4 = re.match(pattern, input4)

 print(match3.group())
 print(match4.group())
```

## search

```python
input3 = "123 Natural Language Processing"
input4 = "Natural Language Processing 123"
```

```python
 pattern = re.compile(r"\d")
 search3 = re.search(pattern, input3)
 search4 = re.search(pattern, input4)

 print(search3.group())
 print(search4.group())
```

# String Substitution and Modification

The rule is used to find matched in the target string and replace them with the specified string. You can specify a maximum number of substitutions, otherwise all matched strings will be replaced.
- sub(rule, replace, target, [count])
- subn(rule, replace, target, [count])


The firest argument is the regex rule, the second is the specified string to replace, the third is the target string, an the fourth is the maximum number of substitutions.

- sub returns a replaced string
- subn returns a tuple with string that was replaced as the first element and a number indicating how many subsitutions were made as the second element.

```python
input3 = "123 Natural Language Processing"
input4 = "Natural Language Processing 123"
```

```python
pattern = re.compile(r"\d")

re.sub(pattern, "num", input3)
```

```python
re.subn(pattern, "num", input4)
```

# Split slicing function

Using the specified regex rule to find matching strings in the target string. Slice the string using these as delimiters:
- split(rule, target, [maxsplit])

The first argument is the regex rule, the second is the target string, and the third is the maxinum number of slices, returning a list of all chopped substrings.

```python
input5 = "NLP123ML456DL"
pattern = re.compile(r"\d+")
re.split(pattern, input5)
```

# (?P<...>) Named groups

<...> inside is the name you gave the group

```python
pattern = re.compile(r"(?P<wow>\d+)(?P<lol>\D+)")
m = re.search(pattern, input5)
```

```python
m.group("wow")
```

```python
m.group("lol")
```

# Example

Filter numbers: XXX-XXX-XXX

```python
input_text = "number: 338-343-220 afsfdf 111-111-111 ahbvnah 1412-3-142"

# pattern = re.compile(r"(\d\d\d-\d\d\d-\d\d\d)")
pattern = re.compile(r"(\d{3}-\d{3}-\d{3})")
m = re.search(pattern, input_text)
m.group()
```

```python
m = re.findall(pattern, input_text)
m
```

- $ means ending at the end
