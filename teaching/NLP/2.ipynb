1. Write a Regular Expression that matches all the numbers in a string.

Target: There are 5 apples and 3 bananas in the basket.

```python
input1 = "There are 5 apples and 3 bananas in the basket."

pattern1 = re.compile(r"\d+")
matches = re.findall(pattern1, input1)
matches
```

2. Find all the dates in the text

Target: Very important date: 25-12-2023, 01-01-2024.

```python
input2 = "Very important date: 25-12-2023, 01-01-2024."

pattern2 = re.compile(r"\d{2}-\d{2}-\d{4}")
dates = re.findall(pattern2, input2)
dates
```

3. Split log files

Target: INFO: 2023-12-17: This is log 1 \n ERROR: 2023-12-18: This is log 2

```python
input3 = " INFO: 2023-12-17: This is log 1 \n ERROR: 2023-12-18: This is log 2"

pattern3 = re.compile(r" \n ")
slices = re.split(pattern3, input3)
slices
```

4. Extracting URLs

Target: You need to visit https://example/com and http://test.org to get the information. If you have any questions you may go to https://help.com.nlp


```python
input4 = "You need to visit https://example/com and http://test.org to get the information. If you have any questions you may go to https://help.com.nlp"

pattern4 = re.compile(r"https?://\S+")
urls = re.findall(pattern4, input4)
urls
```

5. Write a Regular Expression to remove ALL HTML tags from an HTML string.

Target: <html><body><div>Hello World!</div></body></html>

```python
input5 = "<html><body><div>Hello World!</div></body></html>"
pattern5 = re.compile(r"<[^>]+>")
cleaned_text = re.sub(pattern5, "", input5)
cleaned_text
```

6. Verify the strength of the user's password. At least 6 characters (maximum 16), including at least 1 uppercase letter, 1 lowercase letter, 1 number and 1 special character.

Special character: !@%^*?~

```python
pattern6 = r"^(?=.*\d)(?=.*[A-Z])(?=.*[a-z])(?=.*[!@%^*?~]).{6,16}$"

password1 = "Password123?"    # Matches
password2 = "Password123?asfX~afa"    # Not Matches
password3 = "password123?"    # Not Matches
password4 = "Password123"    # Not Matches

print(bool(re.match(pattern6, password1)))
print(bool(re.match(pattern6, password2)))
print(bool(re.match(pattern6, password3)))
print(bool(re.match(pattern6, password4)))
```

 N-grams are a way to preserve context information as data travels through a pipline

```python
sentence = "Thomas Jefferson began building Monticello at the age of 26."

pattern = re.compile(r"([-\s.,:;!?])+")
tokens = pattern.split(sentence)
print(tokens)
[x for x in tokens if x and x not in "- \t\n.,:;!?"]
```

```python
def tokenize_2grams(text):
  tokens = text.split()
  return [(tokens[i], tokens[i+1]) for i in range(len(tokens) - 1)]

tokenize_2grams(sentence)
```

```python
from nltk.util import ngrams

tokens = sentence.split()
list(ngrams(tokens, 2))
```

```python
list(ngrams(tokens, 3))
```

The ngram function behaves like an iterator, generating one element at a time instead of returning the whole sequence at one. This is very useful in for loop, the generator will load each individual item instead of loading the entire list of items into memory

```python
two_grams = list(ngrams(tokens, 2))
[" ".join(x) for x in two_grams]
```

```python
def step1a(self, word):
  if word.endswith("sses"):
    word = self.replace(word, "sses", "ss")
  elif word.endswith("ies"):
    word = self.replace(word, "ies", "i")
  elif word.endswith("ss"):
    word = self.replace(word, "ss", "s")
  elif word.endswith("s"):
    word = self.replace(word, "s", "")
  return word
```

```python
import pandas as pd

movies = pd.read_csv("/content/movieReviewSnippets_GroundTruth.csv")
movies = pd.DataFrame(movies)
```

```python
movies.head().round(2)
```

```python
movies.describe().round(2)
```

```python
pd.set_option("display.width", 75)
from nltk.tokenize import casual_tokenize
from collections import Counter

bags_of_words = []
for text in movies.text:
  bag = Counter(casual_tokenize(text))
  bags_of_words.append(bag)

df_bows = pd.DataFrame.from_records(bags_of_words)
df_bows = df_bows.fillna(0).astype(int)
df_bows.shape
```

```python
df_bows.head()
```

```python
df_bows.head()[list(bags_of_words[0].keys())]
```

```python
movies.sentiment
```

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb = nb.fit(df_bows, movies.sentiment > 0)
movies["predicted_sentiment"] = nb.predict_proba(df_bows)[:, 1] * 8 - 4
movies["error"] = (movies.predicted_sentiment - movies.sentiment).abs()
movies.error.mean().round(1)
```

```python
movies["sentiment_ispositive"] = (movies.sentiment > 0).astype(int)
movies["predicted_ispositive"] = (movies.predicted_sentiment > 0).astype(int)
movies["""sentiment predicted_sentiment sentiment_ispositive predicted_ispositive""".split()].head(8)
```

```python
(movies.predicted_ispositive == movies.sentiment_ispositive).sum() / len(movies)
```
