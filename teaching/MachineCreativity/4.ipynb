```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 100 # Mean
sigma = 1 # std
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f"N({mu}, {sigma}^2)", color="b")
plt.fill_between(x, y, alpha=0.3)
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 100 # Mean
sigmas = [0.5, 1, 1.5, 2] # std
x = np.linspace(mu - 4*max(sigmas), mu + 4*max(sigmas), 1000)


plt.figure(figsize=(10, 6))
for sigma in sigmas:
  y = norm.pdf(x, mu, sigma)
  plt.plot(x, y, label=f"N({mu}, {sigma}^2)")
  plt.fill_between(x, y, alpha=0.3)

plt.title("Normal Distribution with Different Variances")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import boxcox
from scipy.stats import probplot

np.random.seed(0)
data = np.random.exponential(scale=2, size=1000)
data_boxcox, lambda_opt = boxcox(data)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
probplot(data, dist="norm", plot=ax1)
ax1.set_title("Original Data Q-Q Plot")

probplot(data_boxcox, dist="norm", plot=ax2)
ax2.set_title(f"BOX-COX Transformaed Data Q-Q Plot (lambda={lambda_opt:.2f})")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import yeojohnson
from scipy.stats import probplot

np.random.seed(0)
data = np.random.exponential(scale=2, size=1000) - 1
data_yj, lambda_opt = yeojohnson(data)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
probplot(data, dist="norm", plot=ax1)
ax1.set_title("Original Data Q-Q Plot")

probplot(data_yj, dist="norm", plot=ax2)
ax2.set_title(f"Yeo-Johnson Transformaed Data Q-Q Plot (lambda={lambda_opt:.2f})")
plt.tight_layout()
plt.show()
```

# PCA Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.data
y = digits.target

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap="viridis", alpha=0.7)
plt.colorbar(scatter, label="Digit Label")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("2D PCA of MNIST Digits Dataset")
plt.show()
```

# T-SNE

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.data
y = digits.target

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap="viridis", alpha=0.7)
plt.colorbar(scatter, label="Digit Label")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("2D t-SNE of MNIST Digits Dataset")
plt.show()
```
